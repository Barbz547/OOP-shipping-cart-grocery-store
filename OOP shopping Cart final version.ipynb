```python
from IPython.display import clear_output as co

class Cart:
    def __init__(self):
        self.list =[]
        
    def add_item(self, item):
        self.list.append(item)
        
    def remove_item(self, item_str):
        for obj in self.list:
            if item_str in [obj.name for obj in self.list]:
                self.list.remove(obj)
            break
    
    def show_items(self):
        total = 0
        print('=' * 40)
        if not self.list:
            print('You have no items in your cart. ')
        else:
            display_cart = []
            for p in self.list:
                if p.name not in [obj['name'] for obj in display_cart]:
                    display_cart.append({
                        'name': p.name,
                        'price': p.price
                    })
                else:
                    for item in display_cart:
                        if item['name'] == p.name:
                            item['price'] += p.price
            for idx, item in enumerate(display_cart):
                print(f"{idx+1}: {item['name']} [{[obj.name for obj in self.list].count(item['name'])}] ${item['price']}")
        print('=' * 40)
        
        total = float(sum([p.price for p in self.list]))
        print(f"Total: ${total:,.2f}")
    
    def clear_items(self):
        self.list.clear()
    
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    def __repr__(self):
        return f'<Item: {self.name} @${self.price}>'
        
class ShoppingCart:
    @classmethod
    def show_instructions(self):
        print("""Type 'add' to add items to your cart.
Type 'remove' to remove an item from your cart.
Type 'clear' to remove all items from your cart.
Type'show" show all the items in your cart.
Type 'quit' to exit the program.""")
        
    @classmethod
    def run(self):
        cart = Cart()
            
        done = False
        while not done:
            co()
            ShoppingCart.show_instructions()
            cart.show_items()
                
            ask = input("What would you like to do? ").lower()
                
            if ask =='add':
                item_name = input("What item would you like to add?")
                item_price = float(input("How much does the item cost?"))
                cart_item = Item(item_name, item_price)
                cart.add_item(cart_item)
            elif ask == 'remove':
                cart_item= input("What item would you like to remove?")
                cart.remove_item(cart_item)
            elif ask == 'clear':
                confirm = input("Are you sure? This action cannot be undone. Y/N ").lower()
                if confirm == 'y':
                    cart.clear_items()
                elif confirm == 'n':
                    continue
                else:
                    print("That's an invalid command. Try again. ")
                input("Press any key to continue. ")
            elif ask == 'quit':
                confirmed = False

                while not confirmed:
                    confirm = input("Are you sure? Y/N").lower()
                    if confirm == 'y':
                        done= True
                        confirmed = True
                    elif confirm == 'n':
                        break
                    else:
                        print('Invalid comman. Try again. ')
                        input('Press any key to continue. ')
                            
ShoppingCart.run()
```
